```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#coulmn Open converted into numpy array
training_set = data.iloc[:,2:3].values ##training the model, am gonna leave the last values for the prediction 
training_set ##printing the training set
```

```python
len(training_set)
```

```python
def df_to_XY(data,window_size=10): ##function to train x and y by using numpy
 X_train=[]
 y_train=[]

 for i in range(10,len(training_set)):
    X_train.append(training_set[i-10:i,0])
    y_train.append(training_set[i,0])
    
 X_train, y_train = np.array(X_train), np.array(y_train)
 return X_train, y_train
```

```python
WINDOW = 10                  ##declaring WINDOW
X,y = df_to_XY(data,WINDOW)
print(len(X),len(y))
X_train = X[:800]
y_train = y[:800]
X_val = X[800:1000]
y_val = y[800:1000]
X_test = X[1000:]
x_test = y[1000:]
```

```python
#Reshaping(To add new dimensions)
X_train = np.reshape(X_train,(X_train.shape[0],X_train.shape[1],1))
X_val = np.reshape(X_val,(X_val.shape[0],X_val.shape[1],1))
X_test = np.reshape(X_test,(X_test.shape[0],X_test.shape[1],1))
```

```python
#Building the RNN (Recurrent neural network)
#(Recurrent Neural network)is a perticular form of artificial neural network that does well on sequence of data.
#RNN do not need stationaary data, they can perfom on non stationary data as well.,b/c RNN can learn complex pattern in our data
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout #LSTM is a form of RNN (most widely used network for purpose of time series forecasting)
```

```python
regressor = Sequential()
```

```python
#Adding the first LSTM layer with the 50 neurons and some Dropout regularisation
regressor.add(LSTM(units=50, return_sequences = True, input_shape=(X_train.shape[1], 1)))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units=50, return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units=50, return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units=50))
regressor.add(Dropout(0.2))

#Output layer for the prediction 
regressor.add(Dense(units=1))
```

```python
#Compiling
regressor.compile(optimizer='adam',loss='mean_squared_error')
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint,EarlyStopping
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam
```

```python
#fitting the rnn to the training set
history=regressor.fit(X_train,y_train,validation_data=(X_val,y_val),epochs=100, batch_size=32) 
#we called the regressor as input and output and am gonna run it over 100 epochs.
```

```python
his = pd.DataFrame(history.history)
```

```python
his.head()
```

```python
import seaborn as sns
his.columns
history_loss = his[['loss', 'val_loss']] #plotting the loss, val_loss per epoch, which model stores every epoch


fig,axes = plt.subplots(2,1,figsize=(14,8))
plt.subplot(2,1,1)
plt.title("Loss & Val Loss")
sns.lineplot(history_loss,palette="flare");
```

```python
train_pred = regressor.predict(X_train).flatten()
val_pred = regressor.predict(X_val).flatten()
test_pred = regressor.predict(X_test).flatten()
```

```python
#so once model training is completed am gonna be plotting the predicted and non predicted data
pred = np.concatenate([train_pred,val_pred,test_pred])
data_pred = pd.DataFrame(data["temp_max"].copy())
data_pred.columns=["actual"]
data_pred = data_pred[WINDOW:]
data_pred["predicted"] = pred

fig,axes = plt.subplots(2,1,figsize=(14,8),dpi=400)
plt.subplot(2,1,1)
plt.title("Validation Results")
sns.lineplot(data_pred[800:],alpha=0.8,palette="flare",linestyle=None);

plt.subplot(2,1,2)
plt.title("Test Results")
sns.lineplot(data_pred[1000:],alpha=0.8,palette="flare",linestyle=None);
```
